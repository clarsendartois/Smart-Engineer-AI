### üìä **2.4 Data Transformation**

</div>

* **Feature Scaling:** Normalize pixel values (e.g., to [0, 1] or [-1, 1]). numerical features.
* **Encoding Categorical Variables:**  Labels are already integers, often converted to one-hot for classification.
* **Feature Engineering:** Image augmentations (e.g., rotation, flip, zoom) are crucial for deep learning to prevent overfitting. Resizing images.

#### **Your Project Code & Details:**

```python
# Define target image size and batch size
IMG_SIZE = (160, 160) # Standard input size for many pre-trained models (e.g., MobileNetV2, VGG16)
BATCH_SIZE = 32

# Preprocessing function for training data (includes augmentation)
def preprocess_train(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = tf.cast(image, tf.float32) / 255.0 # Normalize pixel values to [0, 1]
    image = tf.image.random_flip_left_right(image)
    # image = tf.image.random_zoom(image, zoom_factor=(0.8, 1.2)) # Example augmentation
    # Replacing random_zoom with random_crop and resize to achieve a similar effect
    # Determine crop size based on zoom_factor (e.g., 0.8 means crop to 80% of size)
    zoom_min = 0.8
    zoom_max = 1.2
    # Calculate the crop size such that resizing it back to IMG_SIZE simulates zoom
    # For zoom_factor z, the crop needs to be 1/z of the original size
    # So for zoom_factor (0.8, 1.2), the crop percentage is (1/1.2, 1/0.8) ‚âà (0.83, 1.25)
    # We want a *random crop* of a smaller size and then resize it back to IMG_SIZE
    # Let's simplify for now and use random_crop with a fixed size and then resize
    # A zoom of 0.8 means the image is scaled up, so we're cropping a smaller area
    # A zoom of 1.2 means the image is scaled down, so we're cropping a larger area (which isn't how random_crop works)
    # A better simulation of random_zoom using available functions involves random crops
    # If zoom_factor is (a, b) where a < 1 and b > 1:
    # If random_zoom < 1, it zooms in (crops a smaller area)
    # If random_zoom > 1, it zooms out (crops a larger area - not possible with random_crop)
    # A simpler approach for random zooming *in* (zoom_factor < 1 for effect)
    # is to randomly crop a fraction of the image and then resize it back.
    # Let's implement zoom-in augmentation (equivalent to zoom_factor < 1)
    scale_factor = tf.random.uniform([], zoom_min, 1.0) # Random scale between zoom_min and 1
    new_height = tf.cast(tf.cast(IMG_SIZE[0], tf.float32) * scale_factor, tf.int32)
    new_width = tf.cast(tf.cast(IMG_SIZE[1], tf.float32) * scale_factor, tf.int32)
    # Ensure crop size is not larger than the image size after initial resize
    crop_height = tf.minimum(new_height, IMG_SIZE[0])
    crop_width = tf.minimum(new_width, IMG_SIZE[1])
    image = tf.image.random_crop(image, size=[crop_height, crop_width, 3])
    image = tf.image.resize(image, IMG_SIZE) # Resize back to original target size


    image = tf.image.random_brightness(image, max_delta=0.2)
    # label = tf.one_hot(label, num_classes) # One-hot encode labels if loss function requires it
    return image, label

# Preprocessing function for validation/test data (no augmentation)
def preprocess_val_test(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = tf.cast(image, tf.float32) / 255.0 # Normalize pixel values to [0, 1]
    # label = tf.one_hot(label, num_classes) # One-hot encode labels
    return image, label

# Apply transformations and create batched datasets
AUTOTUNE = tf.data.AUTOTUNE

ds_train = ds_train_raw.map(preprocess_train, num_parallel_calls=AUTOTUNE)
ds_train = ds_train.cache().shuffle(buffer_size=1000).batch(BATCH_SIZE).prefetch(buffer_size=AUTOTUNE)

ds_val = ds_val_raw.map(preprocess_val_test, num_parallel_calls=AUTOTUNE)
ds_val = ds_val.cache().batch(BATCH_SIZE).prefetch(buffer_size=AUTOTUNE)

ds_test = ds_test_raw.map(preprocess_val_test, num_parallel_calls=AUTOTUNE)
ds_test = ds_test.cache().batch(BATCH_SIZE).prefetch(buffer_size=AUTOTUNE)

print("\nData transformation setup complete. Datasets are batched and preprocessed.")
```

* **Transformation Summary:**
  * Image Resizing: All images resized to 224
times224 pixels.
  * Pixel Normalization: Pixel values scaled to the range [0,1].
  * Image Augmentation (Training): Applied random horizontal flips, random zoom, and random brightness changes to increase data diversity and prevent overfitting.
  * Label Encoding: Labels are integers; will be implicitly handled by SparseCategoricalCrossentropy loss. If using CategoricalCrossentropy, labels would be one-hot encoded.

### üß† **4.2 Train Models**

</div>

* **Fit chosen algorithms to the training data.**

#### **Your Project Code & Details:**

```python
import tensorflow as tf
from tensorflow.keras import layers, models, applications
import tensorflow_datasets as tfds

# Define the base model for transfer learning
# Using MobileNetV2 as a good balance of performance and efficiency
# include_top=False removes the classification head, allowing us to add our own.
# pooling='avg' adds a GlobalAveragePooling2D layer, which is common for classification.
base_model = applications.MobileNetV2(input_shape=IMG_SIZE + (3,),
                                       include_top=False,
                                       weights='imagenet',
                                       pooling='avg',
                                       alpha=0.75  # Smaller model for speed (0.35‚Äì1.0)
                                       ) # Add GlobalAveragePooling2D

# Freeze the base model layers (initial training phase)
base_model.trainable = False

# Build the classification head
inputs = tf.keras.Input(shape=IMG_SIZE + (3,))
x = base_model(inputs, training=False) # Important for batch_norm layers in base_model
x = layers.Dropout(0.2)(x) # Add dropout for regularization
outputs = layers.Dense(num_classes, activation='softmax')(x) # Output layer for multi-class classification

model = models.Model(inputs, outputs)

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # Use sparse_categorical_crossentropy if labels are integers
              metrics=['accuracy',
                        tf.keras.metrics.SparseTopKCategoricalAccuracy(k=5, name='top_5_accuracy')])

model.summary()

# Train the model (initial phase - fine-tuning only the top layers)
print("\nTraining the model (fine-tuning classification head only)...")
initial_epochs = 5  # Reduce for speed
history_initial = model.fit(ds_train,
                            epochs=initial_epochs,
                            validation_data=ds_val)

# --- Fine-tuning phase (unfreeze some layers of the base model) ---
print("\nStarting fine-tuning of the base model...")
base_model.trainable = True # Unfreeze base model

# Freeze all layers except for the last few blocks in the base model
# This strategy is common to avoid destroying early feature detectors.
# Example: Unfreeze layers from a specific layer onwards (e.g., last ~20 layers)
fine_tune_at = len(base_model.layers) - 30 # Adjust based on model architecture, MobileNetV2 has ~150 layers
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Recompile the model with a lower learning rate for fine-tuning
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), # Smaller learning rate
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy',
                        tf.keras.metrics.SparseTopKCategoricalAccuracy(k=5, name='top_5_accuracy')])

fine_tune_epochs = 5
total_epochs = initial_epochs + fine_tune_epochs

history_fine_tune = model.fit(ds_train,
                              epochs=total_epochs,
                              initial_epoch=history_initial.epoch[-1],
                              validation_data=ds_val)

print("\nModel training complete.")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, applications, regularizers
import tensorflow_datasets as tfds

# Constants (set these appropriately)
IMG_SIZE = (160, 160)  # Example image size
num_classes = 10       # Replace with your actual number of classes

# Data Augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomBrightness(0.1),
])

# Apply augmentation only on training set
ds_train = ds_train.map(lambda x, y: (data_augmentation(x), y),
                        num_parallel_calls=tf.data.AUTOTUNE)
ds_train = ds_train.prefetch(tf.data.AUTOTUNE)
ds_val = ds_val.prefetch(tf.data.AUTOTUNE)

# Base model for transfer learning
base_model = applications.MobileNetV2(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights='imagenet',
    pooling='avg',
    alpha=0.75
)
base_model.trainable = False

# Classification head
inputs = tf.keras.Input(shape=IMG_SIZE + (3,))
x = base_model(inputs, training=False)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(num_classes, activation='softmax',
                       kernel_regularizer=regularizers.l2(0.001))(x)

model = models.Model(inputs, outputs)

# Compile initial model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.SparseTopKCategoricalAccuracy(k=5)]
)

model.summary()

# Callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint("best_model.h5", save_best_only=True)
]

# Initial training (only classification head)
print("\nTraining the model (fine-tuning classification head only)...")
initial_epochs = 5
history_initial = model.fit(ds_train,
                            epochs=initial_epochs,
                            validation_data=ds_val,
                            callbacks=callbacks)

# Fine-tuning phase (unfreeze last few layers)
print("\nStarting fine-tuning of the base model...")
base_model.trainable = True

# Unfreeze only the last 10 layers
fine_tune_at = len(base_model.layers) - 10
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Recompile with a lower LR for fine-tuning
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=5e-6),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.SparseTopKCategoricalAccuracy(k=5)]
)

fine_tune_epochs = 5
total_epochs = initial_epochs + fine_tune_epochs

history_fine_tune = model.fit(ds_train,
                              epochs=total_epochs,
                              initial_epoch=history_initial.epoch[-1],
                              validation_data=ds_val,
                              callbacks=callbacks)

print("\nModel training complete.")
```
