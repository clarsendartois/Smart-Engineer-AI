```python
import chess.pgn
import chess
import pandas as pd
import io
import requests
import datetime


def get_games(username, start_year, start_month):
    current_date = datetime.datetime.now()
    games = []

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    for year in range(start_year, current_date.year + 1):
        for month in range(1, 13):
            if year == start_year and month < start_month:
                continue
            if year == current_date.year and month > current_date.month:
                break

            url = f"https://api.chess.com/pub/player/{username}/games/{year}/{month:02d}"
            response = requests.get(url, headers=headers)

            if response.status_code == 200:
                data = response.json()
                if "games" in data and data["games"]:
                    games.extend(data["games"])
                else:
                    print(f"No games in {year}-{month:02d}")
            else:
                print(f"Error fetching data for {year}-{month:02d} (status: {response.status_code})")

    return games

def parse_and_extract_features(games):
    piece_values = {
        chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 3,
        chess.ROOK: 5, chess.QUEEN: 9, chess.KING: 0
    }

    game_data = []

    for g in games:
        pgn_text = g.get("pgn")
        if not pgn_text:
            continue

        game = chess.pgn.read_game(io.StringIO(pgn_text))
        if not game:
            continue

        board = game.board()
        for move in game.mainline_moves():
            board.push(move)

        white_pieces = sum(1 for p in board.piece_map().values() if p.color == chess.WHITE)
        black_pieces = sum(1 for p in board.piece_map().values() if p.color == chess.BLACK)

        material_advantage_white = 0
        for piece in board.piece_map().values():
            value = piece_values.get(piece.piece_type, 0)
            material_advantage_white += value if piece.color == chess.WHITE else -value

        result = game.headers.get('Result', '*')

        game_data.append({
            'Event': game.headers.get('Event'),
            'Site': game.headers.get('Site'),
            'Date': game.headers.get('Date'),
            'Round': game.headers.get('Round'),
            'White': game.headers.get('White'),
            'Black': game.headers.get('Black'),
            'WhiteElo': game.headers.get('WhiteElo'),
            'BlackElo': game.headers.get('BlackElo'),
            'Result': result,
            'WhiteRatingDiff': game.headers.get('WhiteRatingDiff'),
            'BlackRatingDiff': game.headers.get('BlackRatingDiff'),
            'ECO': game.headers.get('ECO'),
            'Opening': game.headers.get('Opening'),
            'TimeControl': game.headers.get('TimeControl'),
            'Termination': game.headers.get('Termination'),
            'WhitePieces': white_pieces,
            'BlackPieces': black_pieces,
            'MaterialAdvantageWhite': material_advantage_white,
            'MoveCount': len(list(game.mainline_moves()))
        })

    return pd.DataFrame(game_data)

# Replace with any Chess.com username
username = "ClarDar"
start_year = 2020
start_month = 3

print(f"Fetching games for {username} starting from {start_year}-{start_month:02d}...")
games = get_games(username, start_year, start_month)
print(f"Retrieved {len(games)} games.")

df_raw = parse_and_extract_features(games)
print("Raw data loaded successfully. Shape:", df_raw.shape)
print("Sample extracted data:")
df_raw.head()


```

### üß† **4.2 Train Models**

</div>

* **Fit chosen algorithms to the training data.**

#### **Your Project Code & Details:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

if 'X_train' in locals() and 'y_train' in locals():
    # Define features
    numerical_features_final = [f for f in ['WhitePieces', 'BlackPieces', 'MaterialAdvantageWhite', 'MoveCount', 'EloDiff', 'AvgElo'] if f in X_train.columns]
    categorical_features_final = [f for f in ['Event', 'Opening', 'TimeControl', 'Termination'] if f in X_train.columns]

    # Preprocessor with remainder=drop to remove raw strings
    preprocessor_final = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_features_final),
            ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features_final)
        ],
        remainder='drop'  # ‚úÖ Important: drop raw columns
    )

    models = {
        'Logistic Regression': Pipeline([
            ('preprocessor', preprocessor_final),
            ('classifier', LogisticRegression(random_state=42, solver='liblinear', multi_class='ovr'))
        ]),
        'Random Forest': Pipeline([
            ('preprocessor', preprocessor_final),
            ('classifier', RandomForestClassifier(random_state=42))
        ]),
        'Gradient Boosting': Pipeline([
            ('preprocessor', preprocessor_final),
            ('classifier', GradientBoostingClassifier(random_state=42))
        ])
    }

    features_final = numerical_features_final + categorical_features_final
    X_train_subset = X_train[features_final]  # ‚úÖ Only selected features

    trained_models = {}
    for name, model in models.items():
        print(f"\nTraining {name}...")
        try:
            model.fit(X_train_subset, y_train)
            trained_models[name] = model
            print(f"{name} trained.")
        except Exception as e:
            print(f"Error training {name}: {e}")
else:
    print("Skipping model training as X_train or y_train was not prepared.")
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

if 'trained_models' in locals() and 'Random Forest' in trained_models:
    # Example: Hyperparameter tuning for Random Forest
    print("\nPerforming Hyperparameter Tuning for Random Forest...")

    # Define parameter grid
    param_grid_rf = {
        'classifier__n_estimators': [100, 200, 300],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 2] # Added for more granular control
    }

    # Use GridSearchCV (can be replaced with RandomizedSearchCV for larger search spaces)
    # Scoring should be appropriate for multiclass, e.g., 'f1_macro'
    grid_search_rf = GridSearchCV(trained_models['Random Forest'], param_grid_rf, cv=3, scoring='f1_macro', n_jobs=-1, verbose=1)
    try:
        grid_search_rf.fit(X_train, y_train)

        print(f"\nBest parameters for Random Forest: {grid_search_rf.best_params_}")
        print(f"Best F1-score (macro) on training set for Random Forest: {grid_search_rf.best_score_:.4f}")

        # Update the trained model with the best estimator
        trained_models['Random Forest (Tuned)'] = grid_search_rf.best_estimator_
    except Exception as e:
        print(f"Error during Random Forest hyperparameter tuning: {e}")

    # Example: Hyperparameter tuning for Logistic Regression
    print("\nPerforming Hyperparameter Tuning for Logistic Regression...")
    param_grid_lr = {
        'classifier__C': [0.01, 0.1, 1, 10, 100], # Inverse of regularization strength
        'classifier__penalty': ['l1', 'l2'] # Regularization penalty
    }
    grid_search_lr = GridSearchCV(trained_models['Logistic Regression'], param_grid_lr, cv=3, scoring='f1_macro', n_jobs=-1, verbose=1)
    try:
        grid_search_lr.fit(X_train, y_train)
        print(f"\nBest parameters for Logistic Regression: {grid_search_lr.best_params_}")
        print(f"Best F1-score (macro) on training set for Logistic Regression: {grid_search_lr.best_score_:.4f}")
        trained_models['Logistic Regression (Tuned)'] = grid_search_lr.best_estimator_
    except Exception as e:
        print(f"Error during Logistic Regression hyperparameter tuning: {e}")
else:
    print("Skipping hyperparameter tuning as models were not trained.")
```

### ‚≠ê **5.1 Evaluate on Validation Set (if applicable)**

```python
# Evaluate all trained models on the validation set
results = {}
if 'trained_models' in locals() and 'X_val' in locals() and 'y_val' in locals():
    for name, model in trained_models.items():
        try:
            y_pred_val = model.predict(X_val)
            # For multiclass, roc_auc_score requires 'ovr' or 'ovo' for 'multi_class'
            # and prediction probabilities. Adjust accordingly.
            # Using 'weighted' or 'macro' for precision, recall, f1

            accuracy = accuracy_score(y_val, y_pred_val)
            precision = precision_score(y_val, y_pred_val, average='macro', zero_division=0)
            recall = recall_score(y_val, y_pred_val, average='macro', zero_division=0)
            f1 = f1_score(y_val, y_pred_val, average='macro', zero_division=0)

            # ROC AUC for multiclass can be complex. For simplicity, we might skip it
            # or calculate per class, or use a specific implementation like one-vs-rest (OvR).
            # If your model supports predict_proba:
            y_prob_val = model.predict_proba(X_val)
            roc_auc = roc_auc_score(y_val, y_prob_val, multi_class='ovr', average='macro')

            results[name] = {
                'Accuracy': accuracy,
                'Precision (Macro)': precision,
                'Recall (Macro)': recall,
                'F1-Score (Macro)': f1,
                'ROC AUC (Macro OvR)': roc_auc
            }
            print(f"\n--- {name} Validation Metrics ---")
            print(f"Accuracy: {accuracy:.4f}")
            print(f"Precision (Macro): {precision:.4f}")
            print(f"Recall (Macro): {recall:.4f}")
            print(f"F1-Score (Macro): {f1:.4f}")
            print(f"ROC AUC (Macro OvR): {roc_auc:.4f}")
        except Exception as e:
            print(f"Error evaluating {name} on validation set: {e}")

    # Convert results to a DataFrame for easy comparison
    results_df = pd.DataFrame(results).T
    print("\nModel Performance on Validation Set:")
    print(results_df)

    # Based on validation set, choose the best model for final evaluation
    # Choose based on F1-Score (Macro) for balanced performance across classes
    if not results_df.empty:
        best_model_name_val = results_df['F1-Score (Macro)'].idxmax()
        final_model = trained_models[best_model_name_val]
        print(f"\nBest model based on validation F1-Score (Macro): {best_model_name_val}")
    else:
        print("No models evaluated successfully on validation set.")
        final_model = None # Ensure final_model is not undefined
else:
    print("Skipping validation evaluation as models or validation data were not prepared.")
```

### ‚≠ê **5.2 Evaluate on Test Set**

</div>

* **Use the completely unseen test set to get an unbiased estimate of the model's generalization performance.**
* **Calculate chosen metrics, generate confusion matrices.**

#### **Your Project Code & Details:**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

if 'final_model' in locals() and final_model is not None and 'X_test' in locals() and 'y_test' in locals():
    print(f"\n--- Evaluating Final Model ({best_model_name_val}) on Test Set ---")

    y_pred_test = final_model.predict(X_test)
    y_prob_test = final_model.predict_proba(X_test) # Probabilities for all classes

    test_accuracy = accuracy_score(y_test, y_pred_test)
    test_precision = precision_score(y_test, y_pred_test, average='macro', zero_division=0)
    test_recall = recall_score(y_test, y_pred_test, average='macro', zero_division=0)
    test_f1 = f1_score(y_test, y_pred_test, average='macro', zero_division=0)
    # Calculate overall macro-averaged OvR AUC
    test_roc_auc = roc_auc_score(y_test, y_prob_test, multi_class='ovr', average='macro')

    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Test Precision (Macro): {test_precision:.4f}")
    print(f"Test Recall (Macro): {test_recall:.4f}")
    print(f"Test F1-Score (Macro): {test_f1:.4f}")
    print(f"Test ROC AUC (Macro OvR): {test_roc_auc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred_test)
    # Map back to original labels for display
    class_labels = ['White Win', 'Black Win', 'Draw']
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)
    plt.figure(figsize=(8, 8))
    disp.plot(cmap=plt.cm.Blues, values_format='d', ax=plt.gca()) # Pass ax to plot on current figure
    plt.title(f'Confusion Matrix for {best_model_name_val} on Test Set')
    plt.show()

    # ROC Curve for multiclass (One-vs-Rest approach)
    plt.figure(figsize=(10, 8))
    # Calculate AUC for each class individually (One-vs-Rest)
    # roc_auc_score with average=None returns AUC for each class
    per_class_roc_auc = roc_auc_score(y_test, y_prob_test, multi_class="ovr", average=None)

    for i, class_label in enumerate(class_labels):
        # roc_curve requires the true labels (binarized for OvR) and the probability estimates for that class
        # To get the true labels binarized for class i: y_test == i
        fpr, tpr, _ = roc_curve(y_test == i, y_prob_test[:, i])
        plt.plot(fpr, tpr, lw=2, label=f'ROC curve for {class_label} (area = {per_class_roc_auc[i]:.2f})') # Use pre-calculated AUC

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic for {best_model_name_val}')
    plt.legend(loc="lower right")
    plt.show()
else:
    print("Skipping test set evaluation as final_model or test data were not prepared.")
```

* **Test Set Performance Summary:** [Final reported metrics and interpretation. e.g., "The Random Forest (Tuned) model achieved a test accuracy of X.XX, with an F1-Score (Macro) of Y.YY. The confusion matrix shows that the model performs relatively well in distinguishing between White wins and Black wins, but might have slightly more difficulty with draws."]
