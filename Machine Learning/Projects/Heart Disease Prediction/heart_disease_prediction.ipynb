* **Feature Engineering:** Create new features from existing ones.

#### **Your Project Code & Details:**

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Identify target variable and features
TARGET_COLUMN = 'target'
X = df_integrated.drop(columns=[TARGET_COLUMN])
y = df_integrated[TARGET_COLUMN]

# Define numerical and categorical features for transformation
# Based on typical understanding of the UCI Heart Disease dataset
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Define preprocessor for numerical and categorical features
# StandardScaler for numerical features (good for distance-based algorithms like SVM, K-NN, Logistic Regression, Neural Networks)
# OneHotEncoder for categorical features (important for nominal categories)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # Keep other columns if any (e.g., 'chol' or 'trestbps' might be numeric too)
)

# Feature Engineering (example - no specific new features for this dataset usually, but for demonstration)
# For example, if we wanted to create age groups:
# X['age_group'] = pd.cut(X['age'], bins=[0, 40, 50, 60, 70, 100], labels=['<40', '40-50', '50-60', '60-70', '>70'])
# This new feature would then be added to categorical_features if numerical_features was adjusted.

print("\nData transformation setup complete. Will be applied within the model pipeline.")
```

* **Divide the dataset into training, validation (optional but recommended), and test sets.**

### 🧠 **4.2 Train Models**

</div>

* **Fit chosen algorithms to the training data.**

#### **Your Project Code & Details:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.pipeline import Pipeline
import xgboost as xgb # For XGBoost

# Ensure numerical_features and categorical_features are correctly identified from X_train
# (Copied from 2.4 to ensure consistency within this block)
numerical_features_final = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
categorical_features_final = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

preprocessor_final = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features_final),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features_final)
    ],
    remainder='passthrough'
)

# Initialize models within pipelines
models = {
    'Logistic Regression': Pipeline(steps=[('preprocessor', preprocessor_final),
                                            ('classifier', LogisticRegression(random_state=42, solver='liblinear', max_iter=1000))]),
    'Random Forest': Pipeline(steps=[('preprocessor', preprocessor_final),
                                      ('classifier', RandomForestClassifier(random_state=42))]),
    'XGBoost': Pipeline(steps=[('preprocessor', preprocessor_final),
                               ('classifier', xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss', use_label_encoder=False, random_state=42))]),
    'SVC': Pipeline(steps=[('preprocessor', preprocessor_final),
                            ('classifier', SVC(probability=True, random_state=42))]) # probability=True for AUC
}

# Train models
trained_models = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    trained_models[name] = model
    print(f"{name} trained.")
```

### 🧠 **4.3 Hyperparameter Tuning**

</div>

* **Optimize model performance by adjusting hyperparameters.**
* **Use techniques like Grid Search, Random Search, or Bayesian Optimization.**

#### **Your Project Code & Details:**

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

print("\nPerforming Hyperparameter Tuning (Example for Random Forest and XGBoost)...")

# --- Random Forest Tuning ---
param_grid_rf = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [5, 10, None],
    'classifier__min_samples_split': [2, 5],
    'classifier__min_samples_leaf': [1, 2]
}

# Using GridSearchCV, optimizing for 'recall' as it's a primary goal
grid_search_rf = GridSearchCV(trained_models['Random Forest'], param_grid_rf, cv=5, scoring='recall', n_jobs=-1, verbose=1)
grid_search_rf.fit(X_train, y_train)

print(f"\nBest parameters for Random Forest: {grid_search_rf.best_params_}")
print(f"Best Recall on training set for Random Forest: {grid_search_rf.best_score_:.4f}")
trained_models['Random Forest (Tuned)'] = grid_search_rf.best_estimator_


# --- XGBoost Tuning ---
param_grid_xgb = {
    'classifier__n_estimators': [100, 200],
    'classifier__learning_rate': [0.01, 0.1, 0.2],
    'classifier__max_depth': [3, 5, 7],
    'classifier__subsample': [0.7, 1.0],
    'classifier__colsample_bytree': [0.7, 1.0]
}

# RandomizedSearchCV for faster exploration if search space is large
random_search_xgb = RandomizedSearchCV(trained_models['XGBoost'], param_grid_xgb, n_iter=20, cv=5, scoring='recall', n_jobs=-1, verbose=1, random_state=42)
random_search_xgb.fit(X_train, y_train)

print(f"\nBest parameters for XGBoost: {random_search_xgb.best_params_}")
print(f"Best Recall on training set for XGBoost: {random_search_xgb.best_score_:.4f}")
trained_models['XGBoost (Tuned)'] = random_search_xgb.best_estimator_

# Update main models dictionary with tuned versions
# trained_models['Random Forest'] = trained_models['Random Forest (Tuned)']
# trained_models['XGBoost'] = trained_models['XGBoost (Tuned)']
```

```python
# Evaluate all trained models (including tuned versions) on the validation set
results = {}
for name, model in trained_models.items():
    if "Tuned" not in name: # Evaluate base models and then tuned ones explicitly if needed
        continue

    y_pred_val = model.predict(X_val)
    y_prob_val = model.predict_proba(X_val)[:, 1] # Probability for the positive class (heart disease)

    accuracy = accuracy_score(y_val, y_pred_val)
    precision = precision_score(y_val, y_pred_val)
    recall = recall_score(y_val, y_pred_val)
    f1 = f1_score(y_val, y_pred_val)
    roc_auc = roc_auc_score(y_val, y_prob_val)

    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC AUC': roc_auc
    }
    print(f"\n--- {name} Validation Metrics ---")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")

# Convert results to a DataFrame for easy comparison
results_df = pd.DataFrame(results).T
print("\nModel Performance on Validation Set:")
print(results_df.sort_values(by='ROC AUC', ascending=False))

# Based on validation set, choose the best model for final evaluation
# Prioritizing Recall and then ROC AUC
best_model_name_val = results_df['Recall'].idxmax() # Start with Recall
# If multiple models have similar recall, use ROC AUC as tie-breaker
# if results_df.loc[best_model_name_val, 'Recall'] == results_df['Recall'].max():
#     best_model_name_val = results_df.loc[results_df['Recall'] == results_df['Recall'].max()]['ROC AUC'].idxmax()

final_model = trained_models[best_model_name_val]
print(f"\nBest model based on validation metrics (prioritizing Recall): {best_model_name_val}")
```
